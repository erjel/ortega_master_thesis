```python
types = ['gaussian','poisson','sp','inpainting','deblurring']

fig,axs = plt.subplots(ncols=3,nrows=len(types),figsize=(15,15))

for t,typ in enumerate(types):

    W = [i.split('_')[-2] for i in glob(f"./checkpoints/lambdas_{typ}_*_1.index")]

    model = get_model("lambdas",10,image_size,num_classes = 1,option = 1)

    E,NE,PSNR,NPSNR = [],[],[],[]

    for w in tqdm(W):
        gen_batch_train,gen_batch_val = get_generators(typ,float(w),CROP1=CROP,BATCH_SIZE=500)
        xs, ys = next(gen_batch_val.as_numpy_iterator())
        edgy = []

        for _ in range(len(ys)):
            im = np.squeeze(ys[_])
            divx = im[1:]-im[:-1]
            divy = im[:,1:]-im[:,:-1]
            edgy.append(np.sum(np.sqrt(divx[:,:-1]**2+divy[:-1]**2)))

        edgy_noise = []

        for _ in range(len(ys)):
            im = np.squeeze(xs[_])
            divx = im[1:]-im[:-1]
            divy = im[:,1:]-im[:,:-1]
            edgy_noise.append(np.sum(np.sqrt(divx[:,:-1]**2+divy[:-1]**2)))



        model.load_weights(f"./checkpoints/lambdas_{typ}_{w}_1")

        p = model(xs)


        psnr = tf.image.psnr(ys,p,1).numpy()
        psnr_noise = tf.image.psnr(ys,xs,1).numpy()


        E.append(np.copy(edgy))
        NE.append(np.copy(edgy_noise))
        PSNR.append(np.copy(psnr))
        NPSNR.append(np.copy(psnr_noise))
    
    
    W = np.asarray(W,dtype='float')
    m = np.array([np.mean(i) for i in NE])
    s = np.array([np.std(i) for i in NE])
    axs[t,0].errorbar(W[np.argsort(W)],m[np.argsort(W)],s[np.argsort(W)])
    axs[t,0].set_xlabel('noise variance')
    axs[t,0].set_ylabel('edginess')
    axs[t,0].set_title(f'Mean edginess of the noisy images\nType of noise: {typ}')
    
    viridis = cm.get_cmap('viridis', 256)
    colors = cm.rainbow(np.linspace(0,1,len(E)))
    newcmp = ListedColormap(colors)
    norm = mpl.colors.Normalize(vmin=np.amin(W), vmax=np.amax(W))
    for i in range(len(E)):

        axs[t,1].scatter(NE[np.argsort(W)[i]],PSNR[np.argsort(W)[i]],s=2,color=colors[i],alpha=0.75)

    plt.colorbar(cm.ScalarMappable(norm=norm, cmap=newcmp),ax=axs[t,1],label='noise variance')
    axs[t,1].set_xlabel('edginess of noisy image')
    axs[t,1].set_ylabel('PSNR')
    axs[t,1].set_title(f'PSNR of denoised image\nvs. edginess of noisy image\nType of noise: {typ}')

    colors = cm.rainbow(np.linspace(0,1,len(E)))
    newcmp = ListedColormap(colors)
    norm = mpl.colors.Normalize(vmin=np.amin(W), vmax=np.amax(W))
    for i in range(len(E)):

        axs[t,2].scatter(E[np.argsort(W)[i]],PSNR[np.argsort(W)[i]],s=2,color=colors[i],alpha=0.75)

    plt.colorbar(cm.ScalarMappable(norm=norm, cmap=newcmp),ax=axs[t,2],label='noise variance')
    axs[t,2].set_xlabel('edginess of original image')
    axs[t,2].set_ylabel('PSNR')
    axs[t,2].set_title(f'PSNR of denoised image\nvs. edginess of image before adding noise\nType of noise: {typ}')

plt.tight_layout()
plt.show()
```

```python
types_noises = ["gaussian","poisson","sp","inpainting","deblurring"]
V = [[2,10],[2,10],[0.05,0.3],[0.05,0.5],[2,10]]
thr = 0
ins = 7

viridis = cm.get_cmap('viridis', 256)
colors = cm.rainbow(np.linspace(0,1,ins))
newcmp = ListedColormap(colors)

fig,axs = plt.subplots(nrows=len(types_noises),ncols=2,figsize=(15,15))

for var in range(len(V)):



    dx,dy,d = [],[],[]
    dxn,dyn,dn = [],[],[]

    for w in tqdm(np.linspace(V[var][0],V[var][1],ins)):
        gen_batch_train,gen_batch_val = get_generators(types_noises[var],w,CROP1=CROP,BATCH_SIZE=50)
        xs, ys = next(gen_batch_train.as_numpy_iterator())

        dx.append([])
        dy.append([])
        d.append([])

        dxn.append([])
        dyn.append([])
        dn.append([])
        for _ in range(len(ys)):
            im = np.copy(np.squeeze(ys[_]))
            divx = im[1:]-im[:-1]
            divy = im[:,1:]-im[:,:-1]
            div = np.sqrt(divx[:,:-1]**2+divy[:-1]**2)

            divx = (np.ndarray.flatten(divx[:,:-1]))
            divy = (np.ndarray.flatten(divy[:-1]))
            div = np.ndarray.flatten(div)

            dx[-1].append(np.copy(divx))
            dy[-1].append(np.copy(divy))
            d[-1].append(np.copy(div))


            im = np.copy(np.squeeze(xs[_]))
            divx_n = im[1:]-im[:-1]
            divy_n = im[:,1:]-im[:,:-1]
            div_n = np.sqrt(divx_n[:,:-1]**2+divy_n[:-1]**2)

            divx_n = (np.ndarray.flatten(divx_n[:,:-1]))
            divy_n = (np.ndarray.flatten(divy_n[:-1]))
            div_n = np.ndarray.flatten(div_n)

            dxn[-1].append(np.copy(divx_n))
            dyn[-1].append(np.copy(divy_n))
            dn[-1].append(np.copy(div_n))

        dx[-1] = np.concatenate(dx[-1])
        dy[-1] = np.concatenate(dy[-1])
        d[-1] = np.concatenate(d[-1])

        dxn[-1] = np.concatenate(dxn[-1])
        dyn[-1] = np.concatenate(dyn[-1])
        dn[-1] = np.concatenate(dn[-1])
        
    if types_noises[var] == "deblurring":
        for i in np.arange(0,ins):

            axs[var,0].scatter(dx[i][d[i]>thr],dy[i][d[i]>thr],s=1,color = colors[i],alpha=0.75)
            axs[var,1].scatter(dxn[i][dn[i]>thr],dyn[i][dn[i]>thr],s=1,color = colors[i],alpha=0.75)
        
    else:
        for i in np.arange(0,ins)[::-1]:

            axs[var,0].scatter(dx[i][d[i]>thr],dy[i][d[i]>thr],s=1,color = colors[i],alpha=0.75)
            axs[var,1].scatter(dxn[i][dn[i]>thr],dyn[i][dn[i]>thr],s=1,color = colors[i],alpha=0.75)
    
    axs[var,0].set_xlabel("/partial x")
    axs[var,0].set_ylabel("/partial y")
    axs[var,0].set_title(f"No added noise")
    axs[var,1].set_xlabel("/partial x")
    axs[var,1].set_ylabel("/partial y")
    axs[var,1].set_title(f"noise: {types_noises[var]}")
    
    norm = mpl.colors.Normalize(vmin=V[var][0], vmax=V[var][1])
    plt.colorbar(cm.ScalarMappable(norm=norm, cmap=newcmp),ax=axs[var,0],label='noise variance')
    plt.colorbar(cm.ScalarMappable(norm=norm, cmap=newcmp),ax=axs[var,1],label='noise variance')
    
    
plt.tight_layout()
plt.show()
```

```python
def divergence_x(x):
    #return(x[:,1:]-x[:,:-1])
    return(x[:,1:-1]-x[:,:-2])

def divergence_y(x):
    #return(x[:,:,1:]-x[:,:,:-1])
    return(x[:,:,1:-1]-x[:,:,:-2])

def divergence_x2(x):
    return(x[:,2:]-x[:,1:-1])

def divergence_y2(x):
    return(x[:,:,2:]-x[:,:,1:-1])

def classifier(inputs, option=1, num_classes=2,kernel_size=3,pool_size=3,CROP=256):
    
    
    x = tf.keras.layers.Conv2D(32, kernel_size, strides=2, padding="same")(inputs)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("relu")(x)

    x = tf.keras.layers.Conv2D(64, kernel_size, padding="same")(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("relu")(x)

    previous_block_activation = x  # Set aside residual

    for size in [16,32,64,128]:
        x = tf.keras.layers.Activation("relu")(x)
        x = tf.keras.layers.SeparableConv2D(size, kernel_size, padding="same")(x)
        x = tf.keras.layers.BatchNormalization()(x)

        x = tf.keras.layers.Activation("relu")(x)
        x = tf.keras.layers.SeparableConv2D(size, kernel_size, padding="same")(x)
        x = tf.keras.layers.BatchNormalization()(x)

        x = tf.keras.layers.MaxPooling2D(pool_size, strides=2, padding="same")(x)

        # Project residual
        residual = tf.keras.layers.Conv2D(size, 1, strides=2, padding="same")(
            previous_block_activation
        )
        x = tf.keras.layers.add([x, residual])  # Add back residual
        previous_block_activation = x  # Set aside next residual

    x = tf.keras.layers.SeparableConv2D(1024, kernel_size, padding="same")(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("relu")(x)

    x = tf.keras.layers.GlobalAveragePooling2D()(x)
    
    return x
```

```python
input_shape = image_size + (1,)
option = 1
num_classes = 1
kernel_size= 3
pool_size =3
it_lim = 10


inputs = tf.keras.Input(shape=input_shape)
x = classifier(inputs, option=option, num_classes=num_classes,kernel_size=kernel_size,
               pool_size=pool_size,CROP=CROP)

x = tf.keras.layers.Dropout(0.5)(x)
lambda_value_1 = tf.keras.layers.Dense(1, activation='linear', name=f'lambdas',kernel_initializer=tf.keras.initializers.RandomNormal(
mean=0.0, stddev=1, seed=None),bias_initializer = tf.keras.initializers.RandomNormal(
mean=0.0, stddev=1, seed=None))(x)
lambda_value_2 = tf.keras.layers.Dense(1, activation='linear', name=f'lambdas',kernel_initializer=tf.keras.initializers.RandomNormal(
mean=0.0, stddev=1, seed=None),bias_initializer = tf.keras.initializers.RandomNormal(
mean=0.0, stddev=1, seed=None))(x)
lambda_value_1 = tf.keras.layers.Lambda(lambda z: tf.math.pow(z,-2))(lambda_value_1)
lambda_value_2 = tf.keras.layers.Lambda(lambda z: tf.math.pow(z,-2))(lambda_value_2)


lam_1 = tf.expand_dims(lambda_value_1,axis=-1)
lam_1 = tf.expand_dims(lam_1,axis=-1)
lam_2 = tf.expand_dims(lambda_value_2,axis=-1)
lam_2 = tf.expand_dims(lam_2,axis=-1)

class_1 = tf.cast(border_classifier(inputs)>0.5,'float32')
class_2 = tf.cast(border_classifier(inputs)<=0.5,'float32')
class_1 = tf.keras.layers.multiply([lam_1,class_1])
class_2 = tf.keras.layers.multiply([lam_2,class_2])

lambda_value = tf.keras.layers.add([class_1,class_2])

outputs = inputs
    
    
for num_it in range(it_lim):

    dx = tf.keras.layers.Lambda(divergence_x)(outputs)
    dy = tf.keras.layers.Lambda(divergence_y)(outputs)

    dx_a = tf.keras.layers.Lambda(divergence_x2)(outputs)
    dy_a = tf.keras.layers.Lambda(divergence_y2)(outputs)

    dx = tf.keras.layers.add([dx,dx_a])
    dy = tf.keras.layers.add([dy,dy_a])

    dx = tf.keras.layers.Lambda(lambda z:tf.image.resize_with_pad(z,CROP,CROP))(dx)
    dy = tf.keras.layers.Lambda(lambda z:tf.image.resize_with_pad(z,CROP,CROP))(dy)

    dx2 = tf.keras.layers.Lambda(lambda z:tf.math.pow(z,2))(dx)
    dy2 = tf.keras.layers.Lambda(lambda z:tf.math.pow(z,2))(dy)

    norm = tf.keras.layers.add([dx2,dy2])
    dx2 = tf.keras.layers.multiply([lambda_value,dx2])
    dy2 = tf.keras.layers.multiply([lambda_value,dy2])

    if option == 1:
        coeff_x = tf.keras.layers.Lambda(lambda z: tf.math.exp(-z),name=f'coeffx_{num_it}')(dx2)
        coeff_y = tf.keras.layers.Lambda(lambda z: tf.math.exp(-z),name=f'coeffy_{num_it}')(dy2)
    elif option == 2:
        coeff_x = tf.keras.layers.Lambda(lambda z: 1./(1. + z),name=f'coeffx_{num_it}')(dx2)
        coeff_y = tf.keras.layers.Lambda(lambda z: 1./(1. + z),name=f'coeffy_{num_it}')(dy2)



    outputs_x = tf.keras.layers.multiply([coeff_x,dx])
    outputs_y = tf.keras.layers.multiply([coeff_y,dy])

    outputs_x = tf.keras.layers.Lambda(divergence_x)(outputs_x)
    outputs_y = tf.keras.layers.Lambda(divergence_y)(outputs_y)

    outputs_x = tf.keras.layers.Lambda(lambda z:tf.image.resize_with_pad(z,CROP,CROP))(outputs_x)
    outputs_y = tf.keras.layers.Lambda(lambda z:tf.image.resize_with_pad(z,CROP,CROP))(outputs_y)

    outputs_it = tf.keras.layers.add([outputs_x,outputs_y])
    outputs = tf.keras.layers.add([0.1*outputs_it,outputs])
```
