# Training

```python
class losses:
    def Linfty(y_true,y_pred):
        error = tf.math.log(tf.math.reduce_mean(tf.exp(tf.abs(y_true-y_pred))))
        return error

    def W1infty(y_true,y_pred):
        error = tf.math.log(tf.math.reduce_mean(tf.exp(tf.abs(y_true-y_pred))))

        dx_true,dy_true = tf.image.image_gradients(y_true)
        dx_pred,dy_pred = tf.image.image_gradients(y_pred)

        error += tf.math.log(tf.math.reduce_mean(tf.exp(tf.abs(dx_true-dx_pred))))
        error += tf.math.log(tf.math.reduce_mean(tf.exp(tf.abs(dy_true-dy_pred))))


        return error

    def H1(y_true,y_pred):
        error = tf.math.reduce_mean(tf.pow(y_true-y_pred,2))

        dx_true,dy_true = tf.image.image_gradients(y_true)
        dx_pred,dy_pred = tf.image.image_gradients(y_pred)

        error += tf.math.reduce_mean(tf.pow(dx_true-dx_pred,2))
        error += tf.math.reduce_mean(tf.pow(dy_true-dy_pred,2))
        error = tf.sqrt(error)

        return error

    def L2(y_true,y_pred):
        error = tf.math.reduce_mean(tf.pow(y_true-y_pred,2))  

        return error
    
    def probability(y_true,y_pred):
        error = -tf.image.ssim(y_true,y_pred,1)

        return error


losses_names = ['Linfty','W1infty','H1','L2','probability']

```

```python
gen_batch_train,gen_batch_val = get_generators(typ,var1_d=5,var1_u=55,CROP1=CROP,BATCH_SIZE=10)

for loss in losses_names:
    print(loss)
    order = 1
    for num_classes in [5,10,15,20,50]:
        print('num_classes: ',num_classes)
        

        
        model = get_model("flux",it_lim=3,image_size=image_size,num_classes = num_classes,
                            with_embedding = True, boundary_detector=True)
        model.compile(optimizer=tf.keras.optimizers.Adam(1e-3),
                loss=getattr(losses,loss))

        callbacks = [tf.keras.callbacks.ModelCheckpoint(
        filepath= f"./checkpoints/flux{loss}tt_{typ}_{num_classes}",
        save_weights_only=True,
        verbose = True,
        save_best_only=True),
        tf.keras.callbacks.ReduceLROnPlateau(factor=0.2, verbose=1,patience=5),
        tf.keras.callbacks.TerminateOnNaN()

        ]

        history = model.fit(
            gen_batch_train,
            epochs=50,
            steps_per_epoch=30,
            validation_data=gen_batch_val,
            validation_steps=10,
            shuffle=False,
            use_multiprocessing=True,
            callbacks=callbacks,
            workers=1
        )

        np.save(f'./history/flux{loss}tt_{typ}_{num_classes}.npy',np.array([history.history['loss'],history.history['val_loss']]))
```

```python
losses_names = ['H1','L2']

order = 1
fig,axs = plt.subplots(ncols=2,nrows=5,figsize=(15,20))

w = np.linspace(10,55,10)

for n,num_classes in tqdm(enumerate([5,10,15,20,50])):
    modelstt = []
    for loss in losses_names:
        model = get_model("flux",it_lim=3,image_size=image_size,num_classes = num_classes)
        try:
            model.load_weights(f"./checkpoints/flux{loss}tt_{typ}_{num_classes}")
        except:
            pass
        modelstt.append(model)
        
    modelsff = []
    for loss in losses_names:
        model = get_model("flux",it_lim=3,image_size=image_size,num_classes = num_classes,
                         with_embedding = False, boundary_detector=False)
        try:
            model.load_weights(f"./checkpoints/flux{loss}ff_{typ}_{num_classes}")
        except:
            pass
        modelsff.append(model)
        
    psnr_tt_1,psnr_tt_s_1 = [[] for l in losses_names],[[] for l in losses_names]
    ssim_tt_1,ssim_tt_s_1 = [[] for l in losses_names],[[] for l in losses_names]
    
    psnr_ff_1,psnr_ff_s_1 = [[] for l in losses_names],[[] for l in losses_names]
    ssim_ff_1,ssim_ff_s_1 = [[] for l in losses_names],[[] for l in losses_names]
        

    gen_batch_train,gen_batch_val = get_generators('gaussian',5,55,CROP1=CROP,BATCH_SIZE=100)
    ims = next(gen_batch_val.as_numpy_iterator())

    for l,loss in enumerate(losses_names):

        res = modelstt[l](ims[0])
        psnr = tf.image.psnr(res,ims[1],1).numpy()
        ssim = tf.image.ssim(res,ims[1],1).numpy()
        psnr_tt_1[l] = psnr
        ssim_tt_1[l] = ssim


        res = modelsff[l](ims[0])
        psnr = tf.image.psnr(res,ims[1],1).numpy()
        ssim = tf.image.ssim(res,ims[1],1).numpy()
        psnr_ff_1[l] = psnr
        ssim_ff_1[l] = ssim
            
    for l,loss in enumerate(losses_names):
        axs[n,0].scatter(psnr_ff_1[l],psnr_tt_1[l],label=loss,alpha=0.5)
        
    M = max(np.amax(psnr_ff_1),np.amax(psnr_tt_1))
    m = max(np.amin(psnr_ff_1),np.amin(psnr_tt_1))
    axs[n,0].plot([m,M],[m,M])
        
    for l,loss in enumerate(losses_names):
        axs[n,1].scatter(ssim_ff_1[l],ssim_tt_1[l],label=loss,alpha=0.5)
        
    M = max(np.amax(ssim_ff_1),np.amax(ssim_tt_1))
    m = max(np.amin(ssim_ff_1),np.amin(ssim_tt_1))
    axs[n,1].plot([m,M],[m,M])
        
    axs[n,0].legend()
    axs[n,1].legend()
        
    
plt.tight_layout()
plt.show()
```

```python
#losses_names = ['Linfty','W1infty','H1','L2','probability']
losses_names = ['H1tt','H1ff']


order = 1
fig,axs = plt.subplots(ncols=2,nrows=5,figsize=(15,20))



for n,num_classes in enumerate([5,10,15,20,50]):
    
    models = []
    for loss in losses_names:
        model = get_model("flux",it_lim=3,image_size=image_size,num_classes = num_classes)
        if 'tt' in loss:
            model = get_model("flux",it_lim=3,image_size=image_size,num_classes = num_classes)
        else:
            model = get_model("flux",it_lim=3,image_size=image_size,num_classes = num_classes,
                         with_embedding = False, boundary_detector=False)
        model.load_weights(f"./checkpoints/flux{loss}_{typ}_{num_classes}")
        models.append(model)
    
    w = np.linspace(10,55,10)
    psnr_nn_1,psnr_nn_s_1 = [[] for l in losses_names],[[] for l in losses_names]
    ssim_nn_1,ssim_nn_s_1 = [[] for l in losses_names],[[] for l in losses_names]

    psnr_pm_1,psnr_pm_s_1 = [],[]
    ssim_pm_1,ssim_pm_s_1 = [],[]

    psnr_or_1,psnr_or_s_1 = [],[]
    ssim_or_1,ssim_or_s_1 = [],[]


    for i,m in enumerate(w):
        gen_batch_train,gen_batch_val = get_generators('gaussian',w[i],w[i],CROP1=CROP,BATCH_SIZE=100)
        ims = next(gen_batch_val.as_numpy_iterator())
        
        for l,loss in enumerate(losses_names):
            
        
            res = models[l](ims[0])
            psnr = tf.image.psnr(res,ims[1],1).numpy()
            ssim = tf.image.ssim(res,ims[1],1).numpy()
            psnr_nn_1[l].append(np.mean(psnr))
            psnr_nn_s_1[l].append(np.std(psnr))
            ssim_nn_1[l].append(np.mean(ssim))
            ssim_nn_s_1[l].append(np.std(ssim))
        
        
        pm = np.array([anisodiff(im,niter = 3,lambd=50) for im in ims[0]])
        pm = np.asarray(pm,dtype='float32')
        psnr = [tf.image.psnr(pm[j],ims[1][j],1).numpy() for j in range(len(res))]
        ssim = [tf.image.ssim(pm[j],ims[1][j],1).numpy() for j in range(len(res))]
        psnr_pm_1.append(np.mean(psnr))
        psnr_pm_s_1.append(np.std(psnr))
        ssim_pm_1.append(np.mean(ssim))
        ssim_pm_s_1.append(np.std(ssim))

        psnr = tf.image.psnr(ims[0],ims[1],1).numpy()
        ssim = tf.image.ssim(ims[0],ims[1],1).numpy()
        psnr_or_1.append(np.mean(psnr))
        psnr_or_s_1.append(np.std(psnr))
        ssim_or_1.append(np.mean(ssim))
        ssim_or_s_1.append(np.std(ssim))

    


    for l,loss in enumerate(losses_names):
        axs[n,0].errorbar(w,psnr_nn_1[l],psnr_nn_s_1[l],label=loss,alpha=0.5)

    axs[n,0].errorbar(w,psnr_pm_1,psnr_pm_s_1,label='pm\nfunction 1',alpha=0.5)
    axs[n,0].errorbar(w,psnr_or_1,psnr_or_s_1,label='noisy',alpha=0.5)


    axs[n,0].set_xlabel('variance of noise')
    axs[n,0].set_ylabel('psnr')
    axs[n,0].set_title(f'PSNR as function of variance in noise\nNumber of classes: {num_classes}')
    axs[n,0].legend()


    for l,loss in enumerate(losses_names):
        axs[n,1].errorbar(w,ssim_nn_1[l],ssim_nn_s_1[l],label=loss,alpha=0.5)
    
    axs[n,1].errorbar(w,ssim_pm_1,ssim_pm_s_1,label='pm\nfunction 1',alpha=0.5)
    axs[n,1].errorbar(w,ssim_or_1,ssim_or_s_1,label='noisy',alpha=0.5)


    axs[n,1].set_xlabel('variance of noise')
    axs[n,1].set_ylabel('ssim')
    axs[n,1].set_title(f'SSIM as function of variance in noise\nNumber of classes: {num_classes}')
    axs[n,1].legend()
    
plt.tight_layout()
plt.show()
```
